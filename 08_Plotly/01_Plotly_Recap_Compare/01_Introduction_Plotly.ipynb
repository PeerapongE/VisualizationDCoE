## 1. Matplotlib

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

### Basic Matplotlib

```python
# step-1: plot data 

plt.scatter(df['total_bill'], 
            df['tip'])
         #label = col)

# step-2: Decoration
plt.xlabel('total bill')
plt.ylabel('tips')
plt.title('This is Basic Matplotlib Chart')
plt.legend() # to show legend label

# step-3: Show the plot
plt.show() # to show plot
```

### Matplotlib with multiple lines

```python
# two different time
df['time'].unique()
```

```python
# step-1: plot data 
for time in df['time'].unique():

    plt.scatter(df[df['time'] == time]['total_bill'], 
                df[df['time'] == time]['tip'],
                label = time)

# step-2: Decoration
plt.xlabel('total bill')
plt.ylabel('tips')
plt.title('This is Matplotlib Chart with two labels')
plt.legend() # to show legend label

# step-3: Show the plot
plt.show() # to show plot
```

## 2. Seaborn

```python
import seaborn as sns
```

```python
# plot data
sns.lmplot(x="total_bill",
           y="tip",
           data=df)

# decoration
plt.title('This is Seaborn Plot')

# show data
plt.show()
```

```python
# plot data
sns.lmplot(x='total_bill',
           y='tip',
           hue = 'time',
           data=df)

# decoration
plt.title('This is Seaborn Plot with hue')

# show data
plt.show()
```

## 3. Interactive Plot: Plotly

https://plot.ly/python/line-and-scatter/

### Basic Scatter Plot

```python
import plotly.offline as pyo
import plotly.graph_objs as go
pyo.init_notebook_mode(connected=True) 
```

```python
trace0 = go.Scatter(x = df['total_bill'],
                    y = df['tip'],
                    mode = 'markers', # to plot scatter plot without line
                   )

data = [trace0]
```

```python
layout = go.Layout(title = 'This is Basic Plotly Scatter Plot', # Graph title
                   xaxis = {'title': 'Total Bill'}, # x-axis label
                   yaxis = {'title': 'Tips'}, # y-axis label
                   hovermode ='closest' # handles multiple points landing on the same vertical
)
```

```python
fig = go.Figure(data   = data,
                layout = layout
               )

# Plot offline
pyo.iplot(fig)
```

### Multiple line with hue

```python
trace_lunch  = go.Scatter(x = df[df['time'] == 'Lunch']['total_bill'],
                          y = df[df['time'] == 'Lunch']['tip'],
                          mode = 'markers', # to plot scatter plot without line
                          name = 'lunch'
                         )

trace_dinner = go.Scatter(x = df[df['time'] == 'Dinner']['total_bill'],
                          y = df[df['time'] == 'Dinner']['tip'],
                          mode = 'markers', # to plot scatter plot without line
                          name = 'dinner'
                         )


data = [trace_lunch, trace_dinner]
```

```python
layout = go.Layout(title = 'This is Basic Plotly Scatter Plot', # Graph title
                   xaxis = {'title': 'Total Bill'}, # x-axis label
                   yaxis = {'title': 'Tips'}, # y-axis label
                   hovermode ='closest' # handles multiple points landing on the same vertical
)
```

```python
fig = go.Figure(data   = data,
                layout = layout
               )

# Plot offline
pyo.iplot(fig)
```
